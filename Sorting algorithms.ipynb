### Let's start with bubble sort algorithm 


```python
# the algorithm takes in a list of integers and will produce a list of the same integers sorted
def bubble_sort(nums):
    for i in range (0, len(nums)): 
        swap = False 
        for j in range (0, len(nums) - 1 - i): 
            if nums[j] > nums[j + 1]: 
                nums[j], nums[j + 1] = nums[j + 1], nums[j] 
                swap = True 
            print(nums)
        print(f"swap is: {swap}") 
        if not swap: 
            break
    return nums 
            
bubble_sort([7,1,3,0,2])
```

### How does this algorithm work?

Let's look at how this algorithm works: 
- We have 2 loops an inner loop and an outer loop
- The outer loop runs `n` times (where n is the number of elememts) but if the list gets sorted before it runs n times, the outerloop can exit early. 
- The inner loop will move the largest unsorted number to the right and 'bubble' it out
- At the end of the iteration of each outer loop one element has been sorted (largest unsorted number has been sorted)
- This allows us to run the inner loop one less time than the previous time because we don't need to compare the largest unsorted number with the largest sorted number.
- Take a look above and you will see that the inner loop runs one less time each time
- If the inner loop finishes all its iterations and there was no swaps made, that means that there was no number to bubble out. That can only happen if all the numbers are in the right place
- In that case we don't need to run the algorithm again
- So we stop once one of two things happen
  1) **There was no swap made in the full iteration of the inner loop**
  2) **The outer loop has gone through all numbers.**

### Lets' look at the time complexity of bubble sort 

**Worse case scenario** 

- We are running the outer loop `n` times 
- and then we run the inner loop `n - 1 - i` times
- On the first iteration we run the inner loop `(n - 1)` times, on the second iteration we run the loop `(n - 2)` times, then `(n - 3)` times untill we get to `1`.
- If we wrote that as an equation it would look like this:
- `(n - 1) + (n - 2) + ... + 2 + 1 = O(T)`
- we can also write it like this:
-  `1 + 2 ... + (n - 2) + (n -1) = O(T)`
-  Now lets add the two equations that gives :
-   `(n - 1 + 1) + (n - 2 + 2) + ... + (2 + n - 1) + (1 + n - 1) = 2O(T)`
-   `n + n + ... n + n = 2O(T)`
-   So we have `n * (n - 1) = 2O(T)`
-   So `O(T) = (n * (n - 1 ))/2`
-   So `O(T) = (n^2 - n)/2`
-   We dont care about constants so that is actually `**O(T) = O(n^2)**`

**Best case scenario** 

- In the best case scenario the list is already sorted. 
- We still enter the inner loop and iterate `**n - 1**` times complete one iteration of it only to find out that swap was false and then we will exit the loop
- Thats give us `O(T) = (n - 1)`
- But we don't care about constants so it's actually `**O(T) = O(n)**`

### Next let's look at insertion sort algorithm 


```python
def insertion_sort(nums): 
    for i in range(1, len(nums)): # this line is okay 
        for j in range(i, 0, -1): # this line is also okay
            if nums[j - 1] > nums[j]: 
                nums[j], nums[j - 1 ] = nums[j - 1], nums[j]
                print(f"{nums} j: {j} i: {i}")
            else: 
                print(f"{nums} j: {j} i: {i}")
                break
    return nums 

insertion_sort([1,7,3,0,2])
```

## How does this algorithm work?

- For this algorithm, we break the array into a sorted part and an unsorted part.
- We ‚Äúcut off‚Äù the first number and assume that it is the sorted part of the array.
- We then move on to the first number not in the sorted part and place it in the correct position within the sorted part of the array.
- We use **`i`** to point to the first unsorted element, so we know where the unsorted array begins.
- We use **`j`** to help us place the unsorted element in the correct position.
- `i` will go through `n ‚àí 1` elements (since `i` starts from index 1).
- `j` will compare the unsorted element with the elements in the sorted part of the array one by one and move it left until it reaches the correct location.
- If a swap is made, `j` is reduced to check whether the element being sorted needs to move further left, until either:
  - no swap is needed (meaning it is now in the correct position), in which case we break and move on to the next `i`, or  
  - `j` reaches `0`, indicating that the element is the new smallest element.

### Let's look at the time complexity of insertion sort 

**The worst case scenario**

-  we have a descending array were we have to sort every elemtn of the array in the unsorted part by moving it all the way to the left side of the sorted part of the array
- In that case we will have to run the outer loop `n - 1` times where `n` is the number of elements in the array
- The inner loop during each iteration of `i` will have to compare the unsorted element with all the elements in the sorted part of the array
- For the first iteration that comparison will happen one time.
- For the second iteration we have 2 sorted elements so the inner loop will run two times.
- So we will have `n - 1` terms in our equation and the equation will look like this:
- `O(T) =  1 + 2 + .... + (n - 2) + (n - 1)`
- Let's rewrite the equation like so:
- `O(T) =  (n - 1) + (n - 2) .... + 2 + 1`
- Let's add the two equations
- `2O(T) =  (n - 1 + 1) + (n - 2 + 2).... + (2 + n - 2) + (1 + n - 1)`
- Since we have n -1 terms that simplifies to `2O(T) =  n + n.... + n + n` and we will have the term `n`, `n - 1` times 
- `2O(T) =  n(n -1)`
- `O(T) =  (n(n -1))/2`
-  We don't care about smaller terms or constants so that leaves us with `O(T) =  n^2`

  
**The best case scenario**

- In the best case scenario we have a sorted array.
- In which case the outer loop still has to loop though `n - 1` times to "verify" that we indeed are putting the elements into the sorted part of the array
- the inner loop will run once for each iteration of the outer loop and since no swap is made, it will break.
- As an equation this would mean that we have `O(T) = 1 + 1 + 1 .... + 1` and we would be adding these ones `n - 1` times
- This gives us `O(T) = 1(n - 1)`
- We don't care about constants so we are left with `O(T) = n`

### Next let's look at the selection sort algortithm 

```python
def selection_sort(nums): 
    for i in range (0, len(nums)): 
        smallest_index = i
        for j in range(i + 1, len(nums)): 
            if nums[j] < nums[smallest_index]: 
                smallest_index = j
        temp = nums[i]
        nums[i] = nums[smallest_index]
        nums[smallest_index] = temp
        print(nums)
    return nums

selection_sort([0,7,3,1,2])
```

## Selection Sort Algorithm

### How It Works

- The array is divided into a **sorted part** (on the left) and an **unsorted part** (on the right).  
- Start at the first element (`i = 0`) and assume it is the smallest element in the unsorted part (`smallest_index = i`).  
- Compare this element with all other elements in the unsorted part:  
  - If a smaller element is found, update `smallest_index` to its index.  
- Once the entire unsorted part has been checked, **swap** the element at `i` with the element at `smallest_index`.  
  - This ensures the smallest element is now at the beginning of the unsorted part, expanding the sorted part.  
- Move to the next element (`i = i + 1`) and repeat the process for the remaining unsorted part.  
- Continue until the outer loop completes. At this point, the array is fully sorted.  
- Each iteration **places the smallest element of the unsorted part into its correct position**.  
- The sorted portion of the array grows by one element at a time.  
- Swapping ensures no elements are lost during the process.


### Let's look at the time complexity of selection sort 

**The worst case scenario**

- In this algorithm the worst and best case scenario is the same because the algorithm will always only complete when the outer loop completed iteration.
- The outer loop will run `n` times from `0` to `n - 1`
- We will have `n` terms in our equation to represent the outer loop.
- In the inner loop we will check `n - 1` elements for the first iteration, `n - 2` elements for the second iteration and so on.
- So we will have `O(T) = (n - 1) + (n - 2) + ... + 2 + 1`
- We can also wrtie that as `O(T) = 1 + 2 + ... + (n - 2) + (n - 1)`
- If we add the two `2O(T) = (1 + n - 1) + (2 + n - 2) + ... + (2 + n - 2) + (1 + n - 1)`
- Then we can simplify like so `2O(T) = n + n+ ... + n + n`
- We don't care about constants so we can get rid of that 2 and we are adding `n` `n` times
- That gives us `O(T) = n^2`

### Next let's look at the merge sort algortithm üò¢   ¬†   

```python
def merge_sort(nums): 
    
    if(len(nums) == 1): #This is our base case. 
        return nums # When the arrays length is equal to one we will just return it because it is sorted

    # If the array is not of size 1, we want to split the array in half 
    M = len(nums)//2 # this will be the middle index
    L = nums[:M] # left side of the array will be up to index M (not inclusive) 
    R = nums[M:] # Right side of the array will be from index M (inclusive) to the end of the array

    # We call merge_sort(R) so the code above runs with R which will split R into right and left again
    R = merge_sort(R) 
    # We call merge_sort(L) so the code above runs with L which will split L into right and left again
    L = merge_sort(L)

    # Execution will resume here once the array is completely split and we have arrays of one element 
    # Now we can start merging our lefts and rights
    sorted_array = merge(L, R)
    return sorted_array

def merge(L, R): 
    sorted_array = []
    L_left, R_left = 0, 0 # the left index of both arrays is 0
    L_right, R_right = len(L) - 1, len(R) - 1 # The right index of both arrays is the last index of both arrays

    # Keep merging as long as the left pointer has not exceeded the right pointer for either array
    while(L_left <= L_right and R_left <= R_right): 
        # if left pointer of left array is pointing tosmaller element, put it in sorted_array first
        if L[L_left] < R[R_left]: 
            sorted_array.append(L[L_left])
            L_left+= 1 # push left arrays left pointer forwards
        else: 
            sorted_array.append(R[R_left]) # if left pointer of right array is pointing to smaller element put it in sorted_array first
            R_left+= 1 # push right arrays left pointer forwards

    # if we have elements left in the left array, add them to the sorted array untill there are no more left
    while(L_left <= L_right): 
        sorted_array.append(L[L_left])
        L_left += 1

    #If we have elements left in the right array, add them to the sorted array until there are no more left
    while(R_left <= R_right): 
        sorted_array.append(R[R_left])
        R_left += 1
        
    return sorted_array
    
            
merge_sort([0,7,3,1,2])        
```

## Merge Sort Algorithm

### How It Works

This one is a little complicated because recursion is hard to envision since it is not executed the same way as non recursive algorithms. To understand this make sure you draw out the recursion or look below for the attached drawing of how the recusion would look for the above call.

 


- When you do recusion problems, you need to break down the problem.
- You cannot think of the problem as a whole or you will get confused.
- The first thing to think of is what is the base case?
- We can see from our image above that our base case is when an array has one element, because an array of one element is a sorted array.
-  So the first thing we will write in our function is what to do when we have the base case.
-  Well when we have an arry of one item it is sorted so we can just return that array.
-  Now what about when we dont have the base case?
-  We want to split the array into right and left portions of the array and sort those portions and then merge them
-  So next we split the array into out right and our left portions
-  Then we recursively call merge_sort on those right and left portions to get the sorted right and left portions
-  What do we do once we have the sorted right and left portions?
-  We will merge them
-  Above there is another function merge that merges the two arrays
-  How does it work?
-  It takes in the sorted right section of the array and the sorted left section of the array.
-  Then it sets left and right pointers for the right array and the left array.
-  It will compare the elements at the left of both parts. Which one is smaller?
-  It will put the smaller element in a new array and then increment the left pointer of that arrays section
-  It will continue doing this while both the right and left section have the left pointers less than or equal to the right pointer.
-  Once that condition is false it is still possible that one of the arrays still has elements inside of it.
-  To get all those elements in the new array that we are making we have two while loops. One for the right array and the left array.
-  They will check if left pointer of the arrays are less than or equal to the right pointer and while they are not they will add the element at the left pointer to the new merged and sorted array and they will increment the left pointer.
-  At the end of the merging they will return the array.
-  Instead of trying to envision the algorithm sequentially just think about it case wise. What do we do in this scenario? what do we do once that is completed?

### Let's look at the time complexity of merge sort 

**The worst and best case scenario**

- This is a recursive algorithm and it splits the problem into smaller problems. In terms of the splitting into right and left sections until we hit arrays of length 1 that will happen `O(log n)` times where n is the number of elements.
- Each time we halve the array untill we hit arrays of length 1. This means that we halved the array `log n` times.
- Starting from the bottom layer we merge the arrays to get to the top. At each layer we merge `n` elements.
- So we are doing `n` operations `log(n)` times which gives us `O(T) = nlog(n)`

### Next let's look at the quick sort algortithm 


```python
def quick_sort(nums): 

    # When the array is empty or it has one item it is sorted 
    n = len(nums)
    # These are our base cases, if we hit them then we can return the array
    if n == 1 or n == 0: 
        return nums

    #if we are not at the base case what do we do?
    pivot = [nums[-1]] # the last element of the array becomes our pivot array

    smaller = [] # we make an empty array to add all the elements into that are smaller than the pivot
    larger = [] # we make an empty array to add all the elemnts into that are larger than the pivot
    
    #split the array in two based on the pivot
    # we don't want to check the last element of the array because it it the pivot.
    # If we check it it will always be added to the smaller array and we will reach max recursion depth
    for num in nums[:-1]: 
        if num <= pivot[0]: 
            smaller.append(num) # add number to smaller if its less than ore equal to the element in the pivot
        else: 
            larger.append(num) # add number to larger if its larger than the element in the pivot
            
    # then we want to keep splitting smaller and larger until we hit the base case (repeat the above) 
    smaller = quick_sort(smaller) 
    larger = quick_sort(larger) 

    # Now our tree is drawn out and we will concatenate smaller and larger to get the sorted arrays 
    new_arr = smaller + pivot + larger 
    return new_arr


quick_sort([0,7,3,1,2])                
```

## Quick Sort Algorithm

### How It Works

This is also a recursive algorithm. It is less complicated than merge sort. Look at the image below to see what the recursive tree would look like:




- This is a recursive problem.
- The base case is when we have an array that is empty or an array of size one because that would mean that the arrays are sorted.
- We start of by picking a pivot.
- The pivot in the algorithm above is the last element.
- Then we make an array of all the elements less than or equal to the pivot called `smaller`, all the elements larger than the pivot called `larger` and an array of the pivot calles `pivot`
- When doing the above step we don't look at the last element since it was the pivot. We would end up reaching the max reursive depth because the last element would always end up in the `smaller` and `smaller` would not reduce in size. You can try draw it out if you are confused. 
- We will call quick_sort() with the smaller and larger array to further split the arrays.
- Once we have hit the base case the tree is drawn out.
- Now we want to merge `smaller` + `pivot` + `larger`. since the tree was split out to put the elements in order merging is jsut concatenating the arrays. 

### Let's look at the time complexity of quick sort 

**The average case scenario**

- We have to split the tree untill all the arrays are empty or have one element. This means that the tree will have log n layers.
- For each alyer we are just concatenating the items.
- To concatenate the items we go through them and that is n items per layer
- This gives us `O(T) = nlogn` in the average case secario
 
**The worst case scenario**

- The pivots that we pick matter here.
- If we pick a bad pivot we can get quadratic time complexity. To pick a good pivot, the pivot has to be random so there should be no ordering to the numbers.
