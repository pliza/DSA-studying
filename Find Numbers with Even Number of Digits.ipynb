```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        # need to find the number of numbers that have an even number of digits in the array nums 
        even_digits = 0
        for num in nums: 
            num_counter=0
            while num >= 1: 
                num = num/10 
                num_counter +=1 
            if num_counter%2==0: 
                even_digits += 1
        return even_digits
```

To get the number of even numbered digits we need to go through all of the elements in the array. To get the numebr of digits in each number we can repeatedly divide the number by 10. eg :

So the number of times that we need to divide the digit by 10 to get a number smaller than 1 is the number of digits in the number. 

What does the complextiy of this look like?

Well first of all we are going through the whole array so assuming our array is made up of n elements, we have to go through O(n) items.

For every number in the array we also need to repeatedly divide it by 10 untill the number becomes smaller than 0. We will perform that operation m times assuming that the number of digits is m. That gives us O(m) for every digit in the array.
